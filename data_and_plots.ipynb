# First part: $\Delta$ vs density plot at fixed temperature

**Parameters**: Nz = 10, eps_r = 12.5, m*_parallel = 0.07 m_e, dz = 3 angstrom, k_B T = 0.05 meV

```julia
densityArray = 5*range(1,20,20); #In units of 10^10/cm^2
DeltaArray = [1.265, 1.697, 2.057, 2.324, 2.519, 2.538, 2.606, 2.695, 2.765, 
                2.854, 2.901, 3.029, 2.700, 2.184, 1.815, 1.322, 1.056, 1.044, 0.897, 0.895]; ## meV
```

```julia
plt1 = plot(densityArray, DeltaArray,
    ylims=(0,maximum(DeltaArray)+0.1),
    seriestype = :line, marker = (:circle, 10), color = :blue, framestyle = :box, linewidth = 2, 
    legend = false, xlabel = "n (10^10/cm^2)", ylabel = "|Δ| (meV)", 
    xguidefontsize = 15, yguidefontsize = 15, tickfontsize = 22)

plt1 = scatter!(densityArray, DeltaArray,
    color = :red, markersize = 6)

display(plt1)
```

# Second part: $\Delta$ vs temperature plot at fixed density

**Parameters**: Nz = 10, eps_r = 12.5, m*_parallel = 0.07 m_e, dz = 3 angstrom, n = 55*10^10/cm^2

1 meV = 11.6045 Kelvin * k_B

```julia
kbTArray = [0.05, 0.15, 0.3, 0.45, 0.6, 0.75, 0.9, 1.05, 1.2, 1.35, 1.5, 1.65, 1.75, 1.8]*11.6045; ## Kelvin

DeltavsTArray = [2.92, 2.948, 2.911, 2.833, 2.72, 2.603, 2.458, 
                    2.28, 2.077, 1.819, 1.483, 0.988, 0.3299, 0.026]; ## meV
```

```julia
plt2 = plot(kbTArray, DeltavsTArray,
    ylims=(-0.1,maximum(DeltaArray)+0.1),
    seriestype = :line, marker = (:circle, 10), color = :blue, framestyle = :box, linewidth = 2, 
    legend = false, xlabel = "T (K)", ylabel = "|Δ| (meV)", 
    xguidefontsize = 15, yguidefontsize = 15, tickfontsize = 22)

plt2 = scatter!(kbTArray, DeltavsTArray,
    color = :red, markersize = 6)

display(plt2)
```
